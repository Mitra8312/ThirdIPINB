```python
import pandas as pd
import numpy as np

df = pd.read_csv("vgsales_22.csv")
df
```

За ненадобностью и в целях уменьшения количества строковых значений параметров убираем параметр "Имя" из выборки. Это поле не относится к значимым, потому что оно уникально.

```python
df = df.drop(columns='Name')
df
```

1. Продемонстрировать работу OrdinalEncoder и использовать его на датасете.
Данное преобразование позволяет нам изменить данные в столбцах, которые ранее были типа object на численный. С помощью функции fit() получим данные из столбцов для дальнейшего изменения.

```python
from sklearn.preprocessing import OrdinalEncoder

Oenc = OrdinalEncoder()
Oenc.fit(df[['Genre', 'Platform']])
df
```

Далее уже с помощью fit_transform мы проводим два действия, сначала заполняем OrdinalEncoder данными со столбцов, после трансформируем (изменяем) датафрейм.
OrdinalEncoder работает по такому принципу, что заменяет данные на их номера по порядку, как они были сохранены в массив при использовании функции fit()

```python
Oenc = OrdinalEncoder()
df[['Genre', 'Platform']] = Oenc.fit_transform(df[['Genre', 'Platform']])
df
```

2. Обязательно преобразовать 1 категориальный признак с помощью LabelEncoder.
Данное преобразование позволяет изменить данные типа object в численное, исходя из индекса массива, т.е. может быть от 0 до -1

```python
from sklearn.preprocessing import LabelEncoder

Lenc = LabelEncoder()
Lenc.fit(df['Publisher'])
df['Publisher'] = Lenc.fit_transform(df['Publisher'])
df
```

Показать пример выполнения OneHotEncoder (использовать его на датасете не обязательно). Данное преобразование работает по бинарному признаку, то есть в конечной выборке значения будут только 1 и 0. В моем датасете нет данных, которые подходят, поэтому используем датасет с пары.

```python
from sklearn.preprocessing import OneHotEncoder, LabelEncoder

df1 = pd.read_csv("weatherAUS_clear.csv")

Lenc = LabelEncoder()
Lenc.fit(df1['RainTomorrow'])
df1['RainTomorrow'] = Lenc.fit_transform(df1['RainTomorrow'])

OHenc = OneHotEncoder()
OHenc.fit(df1[['RainToday']])

OHenc.transform(df1[['RainToday']]).toarray()
```

 4. Провести нормализацию с помощью Normalize на числовых признаках;

Сейчас мы проанализируем выборку, чтобы недопустить ошибки при нормализации. Если выборка имеет незаполненные поля, то нормализация не сможет выполниться корректно.

```python
df.info()
```

Тут мы видим, что параметр "Год" имеет 7 незаполненных значений. С помощью функции, которая заполнит пропущенные поля средними, мы выйдем из этой ситуации.

```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy='mean')
df['Year'] = imputer.fit_transform(df[['Year']])
df.info()
```

Теперь незаполненных значений в выборке нет. Можно переходить к нормализации.

```python
from sklearn.preprocessing import Normalizer

Norm = Normalizer()

df[['Rank','Platform', 'Year', 'Genre',	'Publisher', 'NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales', 'Global_Sales']] = Norm.fit_transform(df[['Rank','Platform', 'Year', 'Genre','Publisher', 'NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales', 'Global_Sales']])
df
```

5. Разделить датасет на обучающую и тестовую выборку.

```python
from sklearn.model_selection import train_test_split

Y = df[['Rank', 'Platform', 'Year', 'Genre', 'Publisher']]
X = df.drop(['Rank', 'Platform', 'Year', 'Genre', 'Publisher'], axis=1)

x_train, x_test, y_train, y_test = train_test_split(X,Y,test_size=0.2,random_state=42)
```
